# 2.2 Activation Functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def perceptron(signal: float) -> np.float32:
    if signal >= 0:
        return(np.float32(1))
    else:
        return(np.float32(0))
```

```python
def sigmoid(signal: float) -> np.float32:
    return 1/(1+np.exp(-signal))
```

```python
def relu(signal: float) -> np.float32:
    if signal >= 0:
        return(np.float32(signal))
    else:
        return(np.float32(0))
```

```python
def tanh(signal: float) -> np.float32:
    return np.float32(np.tanh(signal))
```

```python
def plot_activation(activation_func_call):
    x_axis = np.arange(-10, 10, 0.1)
    plt.axis([-10, 10, -3, 3])
    plt.axhline(y=0, color='k')
    plt.axvline(x=0, color='k')
    return plt.plot(x_axis, list(map(activation_func_call, x_axis)))
```

```python
plot_activation(perceptron)
```

```python
plot_activation(sigmoid)
```

```python
plot_activation(relu)
```

```python
plot_activation(tanh)
```
